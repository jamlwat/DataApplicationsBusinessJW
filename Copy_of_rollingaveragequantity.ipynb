```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/camdenlol/SnOasis/main/snoasisData.csv')

df["DateTime"] = pd.to_datetime(df.Date +" "+ df.Time, format = '%m/%d/%Y %I:%M:%S?%p')

df.head()
```

```python
df["dayofyear"]=df.DateTime.dt.day_of_year
```

```python

dfg=df.groupby("dayofyear").Quantity.sum()
dfg
```

```python
dfg[59]
```

```python
rollingaverage = dfg.rolling(7).mean()
rollingaverage
```
